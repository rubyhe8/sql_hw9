### `Batting` column definitions

Each row is one **player–season–stint**: a player can have multiple rows in the same year if they played for more than one team (`stint` orders those stays).

| Column | Meaning |
|--------|---------|
| `playerID` | Player identifier; matches `People.playerID` and keys in other tables. |
| `yearID` | Season year. |
| `stint` | `1`, `2`, … for the first, second, … team that season (same player, same year). |
| `teamID` | Team abbreviation for that stint. |
| `lgID` | League (`AL`, `NL`, etc.). |
| `G` | Games played (team games the player appeared in). |
| `G_batting` | Games in which the player had a batting line (useful for pitchers); often `NULL` in older rows. |
| `AB` | Official **at-bats** (plate appearances that are not walks, HBP, sac hits/flies, or catcher interference). |
| `R` | **Runs** scored. |
| `H` | **Hits** (singles + doubles + triples + home runs). |
| `2B` | **Doubles**. |
| `3B` | **Triples**. |
| `HR` | **Home runs**. |
| `RBI` | **Runs batted in**. |
| `SB` | **Stolen bases**. |
| `CS` | **Caught stealing**. |
| `BB` | **Walks** (bases on balls), excluding intentional walks in modern scoring (`IBB` tracked separately). |
| `SO` | **Strikeouts** (batting). |
| `IBB` | **Intentional walks** (tracked when available). |
| `HBP` | **Hit by pitch**. |
| `SH` | **Sacrifice bunts** (sacrifice hits). |
| `SF` | **Sacrifice flies**. |
| `GIDP` | **Grounded into double play**. |
| `G_old` | Legacy / deprecated games field from older Lahman releases; often `NULL`. |

## Explore the hall of fame table

```python
import pandas as pd

query = """
SELECT * FROM HallOfFame
"""

df = pd.read_sql(query, conn)
df
```

### `HallOfFame` column definitions

Each row is one **Hall of Fame ballot appearance**: the same `playerID` can appear many times across different election years or voting bodies.

| Column | Meaning |
|--------|---------|
| `playerID` | Player (or honoree) identifier; matches `People.playerID` when the row is for a person in the database. |
| `yearid` | **Election year** (calendar year of the vote), not the season being honored. |
| `votedBy` | **Voting body** or election type (e.g. `BBWAA` = Baseball Writers’ Association of America; other codes for veterans, era committees, etc.). |
| `ballots` | Number of **ballots cast** in that election (when recorded). |
| `needed` | **Votes required** for induction in that election (when recorded). |
| `votes` | **Votes received** by this candidate in that election. |
| `inducted` | `Y` if the candidate was **inducted** as a result of this row’s election; `N` otherwise. |
| `category` | **Ballot category**: `Player`, `Manager`, `Pioneer/Executive`, `Umpire`, etc. |
| `needed_note` | Optional **footnote** text from the source data (often `NULL`). |

## Lets try to predict a players induction into the hall of fame based on their batting statistics

## We first query our data

We pull **career batting totals** from `Batting`. The label is **Player** induction in `HallOfFame` (`category = 'Player'` and `inducted = 'Y'`).

**Train / test split**

**Model:** standardized features + **logistic regression** (`class_weight='balanced'` because inductees are rare). Metrics on the held-out era include **ROC-AUC**, accuracy, confusion matrix, and a classification report.

```python
import pandas as pd

hof_query = """
WITH career AS (
  SELECT
    playerID,
    SUM(AB) AS ab,
    SUM(R) AS r,
    SUM(H) AS h,
    SUM("2B") AS x2b,
    SUM("3B") AS x3b,
    SUM(HR) AS hr,
    SUM(RBI) AS rbi,
    SUM(BB) AS bb,
    SUM(SO) AS so,
    SUM(SB) AS sb
  FROM Batting
  GROUP BY playerID
),
hof_player AS (
  SELECT DISTINCT playerID
  FROM HallOfFame
  WHERE category = 'Player' AND inducted = 'Y'
)
SELECT
  c.playerID,
  c.ab,
  c.r,
  c.h,
  c.x2b,
  c.x3b,
  c.hr,
  c.rbi,
  c.bb,
  c.so,
  c.sb,
  CASE WHEN h.playerID IS NOT NULL THEN 1 ELSE 0 END AS inducted
FROM career c
LEFT JOIN hof_player h
ON c.playerID = h.playerID;
"""

players = pd.read_sql(hof_query, conn)
players
```
